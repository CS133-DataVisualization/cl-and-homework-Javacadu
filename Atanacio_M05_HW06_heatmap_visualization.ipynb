And, IGN agrees with her that it is a great game -- their rating for this game is a whopping 8.9! Inspired by the success of this game, your instructor is considering creating your very own racing game for the Wii platform. Perform the following analyses to help her determine which platform she should focus on.

## Problem 5
Use the Pokemon dataset to create a clustermap with color. First, filter the dataframe to only keep data with `Type 1` equals to following values: `Water`, `Normal`, `Grass`, `Bug` and `Psychic`. Annotate the dendrogram using different colors for these five different `Type 1` values. Use `Name` as the index.

pokemon_data is in https://raw.githubusercontent.com/csbfx/advpy122-data/master/Pokemon.csv

```python
pokemon_url = "https://raw.githubusercontent.com/csbfx/advpy122-data/master/Pokemon.csv"
pokemon_df = pd.read_csv(pokemon_url)

# Filter for specific types
types_to_keep = ['Water', 'Normal', 'Grass', 'Bug', 'Psychic']
filtered_pokemon = pokemon_df[pokemon_df['Type 1'].isin(types_to_keep)].copy()

# Prepare data for clustermap
filtered_pokemon = filtered_pokemon.set_index('Name')
data_for_clustermap = filtered_pokemon.select_dtypes(include=['number']).drop(columns=['#', 'Generation', 'Legendary'], errors='ignore')

# Create color mapping for Type 1
type_colors = filtered_pokemon['Type 1'].map({
    'Water': 'blue',
    'Normal': 'gray',
    'Grass': 'green',
    'Bug': 'olive',
    'Psychic': 'pink'
})

sns.clustermap(data_for_clustermap, row_colors=type_colors, standard_scale=1)
plt.show()
```
